```python
import numpy as np
import pandas as pd
```

```python
books= pd.read_csv('./data/books.csv', sep =";", on_bad_lines= 'skip', encoding = 'latin1', )
books.head()
```

```python
books = books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher']]
```

```python
books.rename(columns={'Book-Title':'title', 'Book-Author':'author', 'Year-Of-Publication':'year', 'Publisher':'publisher'}, inplace = True)
books.head(2)
```

```python
users = pd.read_csv('./data/users.csv', sep =";", on_bad_lines='skip', encoding='latin1')
```

```python
users.head()
```

```python
users.rename(columns={'User-ID': 'id',
                     'Location': 'location',
                     'Age': 'age'}, inplace= True)
```

```python
ratings = pd.read_csv('./data/ratings.csv', sep =";", on_bad_lines='skip', encoding='latin1')
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID': 'user_id',
                       'Book-Rating': 'rating'}, inplace = True)
```

* Keep users who have given atleast 200 reviews
* Keep books which has atleast 50 reviews

```python
ratings['user_id'].value_counts().shape
```

```python
x = ratings['user_id'].value_counts()>200
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head()
books.head()
```

```python
ratings_with_books = ratings.merge(books, on = 'ISBN')
```

```python
ratings_with_books.shape
```

```python
number_of_ratings = ratings_with_books.groupby('title')['rating'].count().reset_index()
number_of_ratings.rename(columns={'rating':'number_of_ratings'}, inplace= True)
number_of_ratings
```

```python
final_rating = ratings_with_books.merge(number_of_ratings, on = 'title')
final_rating.head(2)
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['number_of_ratings']>=50]
```

```python
final_rating.drop_duplicates(['user_id', 'title'], inplace= True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values = 'rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[240, :].values.reshape(1, -1), n_neighbors=6)
```

```python
distances
```

```python
suggestions
type(suggestions)
suggestions[0]
```

```python
for i in range(len(suggestions[0])):
    print(i+1, book_pivot.index[suggestions[0][i]])
```

```python
np.where(book_pivot.index == 'Animal Farm')[0][0]
```

```python
def recommend(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)

    for i in range(len(suggestions[0])):
        if i!=0:
         print(book_pivot.index[suggestions[0][i]])
```

```python
recommend('Tough Cookie')
```
